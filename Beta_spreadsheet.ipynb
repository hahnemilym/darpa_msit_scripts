# Create CSV for Beta Vals

```python
import os
import numpy as np
import pandas as pd
from pandas import DataFrame
from pandas import read_csv
from pandas import read_table
# import nibabel as nib
from scipy.signal import detrend
#from mne import read_labels_from_annot
#from mne.filter import high_pass_filter
from sklearn.decomposition import PCA
from IPython.display import display

#~~~~~~~~~~~~~~~~~#
## Set Variables
#~~~~~~~~~~~~~~~~#

beta_dir = '/autofs/space/neuroendo_001/users/AREX/subjects/batch_script/1127/rsfc2017/extractions_112017_indsub/'
subjid = read_csv('subjidall.txt', header=None, names=['subjid'])
hemis = ['lh', 'rh']

for h in hemis:

    betas = read_csv('labels_%s.txt'%h, header=None)
    columns = betas[0]
    
#~~~~~~~~~~~~~~#
## Set up DF
#~~~~~~~~~~~~~~#
    df_active = pd.DataFrame(index=range(0,len(subjid)),columns=columns)
    df_placebo = pd.DataFrame(index=range(0,len(subjid)),columns=columns)
#     display(df)
    
    for beta in betas[0]:
        i=0
        for subject in subjid['subjid']:
            rval = read_table(beta_dir + '%s_%s_avgwf_%s.dat'%(subject,h,beta), header=None)
            rval=rval[0].ix[0]
            if i % 2 == 0: 
                df_active[beta].ix[i]=rval
            else:
                df_placebo[beta].ix[i]=rval
            i+=1
        
    df_active = df_active.dropna(axis=0)
    df_active= df_active.reset_index()
    df_placebo = df_placebo.dropna(axis=0)
    df_placebo= df_placebo.reset_index()
#     display(df_placebo)
#~~~~~~~~~~~~~~~~~~~~~~~~~~~#
## Make a Super Spreadsheet
#~~~~~~~~~~~~~~~~~~~~~~~~~~#

    df_all = pd.DataFrame(index=range(0,10))
    
    for beta in betas[0]:
        df_all[beta+'_active'] = df_active[beta]
        df_all[beta+'_placebo'] = df_placebo[beta]
        df_all[beta+'_A-P'] = df_active[beta] - df_placebo[beta]
    display(df_all)
#~~~~~~~~~~~~#
##Save
#~~~~~~~~~~~#

    df_active.to_csv('%s_active.csv'%h)
    df_placebo.to_csv('%s_placebo.csv'%h)
    df_all.to_csv('%s_tosend.csv'%h)
    
```
